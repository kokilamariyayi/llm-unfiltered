1️⃣ Basic Tokenization with tiktoken

```python
import tiktoken

# Load the encoding used by GPT-4
enc = tiktoken.get_encoding('cl100k_base')

text = 'LLMs Unfiltered — Beyond the Prompt'
tokens = enc.encode(text)

print('Text:', text)
print('Tokens:', tokens)
print('Token count:', len(tokens))
```

2️⃣ Compare Token Counts Across Different Texts

```python
samples = [
    'cat',
    'Unfiltered',
    '12345',
    'Hello world!',
    'The model predicts the next token',
    'transformer',
    't-r-a-n-s-f-o-r-m-e-r',
    '🤖',
]

print(f"{'Text':<40} {'Token Count'}")
print('-' * 55)
for sample in samples:
    count = len(enc.encode(sample))
    print(f"{sample:<40} {count}")
```

3️⃣ The Multilingual Problem

```python
multilingual = [
    ('English', 'The model predicts the next token'),
    ('Hindi',   'मॉडल अगले टोकन की भविष्यवाणी करता है'),
    ('Tamil',   'மாடல் அடுத்த டோக்கனை கணிக்கிறது'),
    ('Arabic',  'يتنبأ النموذج بالرمز التالي'),
]

print(f"{'Language':<12} {'Token Count':<15} {'Text'}")
print('-' * 70)
for lang, text in multilingual:
    count = len(enc.encode(text))
    print(f"{lang:<12} {count:<15} {text}")
```

4️⃣ Decode Tokens Back to Text

```python
text = 'Tokenization is the first step in understanding LLMs'
tokens = enc.encode(text)

print('Original text:', text)
print('\nToken by token breakdown:')
for token in tokens:
    decoded = enc.decode([token])
    print(f'  Token ID: {token:<8} → "{decoded}"')
```

✅ Key Takeaways

- LLMs don't read words — they read **tokens**
- Tokens ≠ words. One word can be multiple tokens
- **BPE** builds vocabulary by merging frequent character pairs
- Non-English languages use **more tokens** for the same meaning
- More tokens = **higher API cost** + faster context window usage
